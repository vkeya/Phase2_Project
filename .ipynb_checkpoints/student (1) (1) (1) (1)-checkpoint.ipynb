<strong>BASELINE MODEL

```python
Y = data['Price']
X = data['Sqft_living']
plt.scatter(X,Y)
plt.xlabel("INDEPENDENT VARIABLES")
plt.ylabel("DEPENDENT VARIABLES")
plt.show;
```

```python
import statsmodels.api as sm
model = sm.OLS(endog=Y, exog=sm.add_constant(X))
model
```

```python
#fitting the model
results = model.fit()
results
```

```python
#checking the p-value
results.fvalue, results.f_pvalue
```

Our model p-value is below 0.0, meaning it is statistically significant.

```python
#Checking our R-squared for our baseline model
results.rsquared
```

The relationship between Price and Sqft_living is linear but not perfectly linear

```python
# Step 2: Split the data into training and test sets
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Step 3: Train the baseline model and calculate RMSE
# Calculate baseline prediction (e.g., using mean of y_train)
baseline_prediction = np.mean(y_train)# Baseline prediction is the mean
# Create an array of baseline predictions for the test set
Y_baseline_pred = np.full_like(y_test, fill_value=baseline_prediction)
# Calculate RMSE for the baseline model
baseline_rmse = np.sqrt(mean_squared_error(y_test, Y_baseline_pred))

print("Baseline RMSE:", baseline_rmse)
```

```python
#find RMSE
from sklearn.metrics import mean_squared_error

# Calculate RMSE
mse = mean_squared_error(y_test, y_pred)  # Calculate Mean Squared Error
rmse = np.sqrt(mse)  # Calculate Root Mean Squared Error

print("RMSE:", rmse)
```
