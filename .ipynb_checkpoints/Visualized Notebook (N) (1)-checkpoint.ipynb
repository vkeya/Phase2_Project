```python
import pandas as pd
import numpy as np


# Columns to check for outliers
columns_to_check = ['bathrooms', 'bedrooms', 'sqft_living', 'sqft_above']

# Calculate z-scores for selected columns
z_scores = np.abs((data[columns_to_check] - data[columns_to_check].mean()) / data[columns_to_check].std())

# Define threshold for outlier detection (e.g., z-score > 3)
threshold = 3

# Identify rows with outliers based on z-scores
outlier_mask = (z_scores > threshold).any(axis=1)

# Filter out rows containing outliers 
data_filtered = data[~outlier_mask]

# Display summary (optional)
print(f"Original DataFrame shape: {data.shape}")
print(f"Filtered DataFrame shape (after removing outliers): {data_filtered.shape}")

# Optionally, you can assign the filtered DataFrame back to 'data' if needed
data = data_filtered
```

### Using the Date Column to come up with a new column called seasons

```python
# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Extract month to determine the season
data['Month'] = data['Date'].dt.month

# Define a function to map month to season
def get_season(month):
    if month in [12, 1, 2]:  # Winter: December, January, February
        return 'Winter'
    elif month in [3, 4, 5]:  # Spring: March, April, May
        return 'Spring'
    elif month in [6, 7, 8]:  # Summer: June, July, August
        return 'Summer'
    else:  # Fall: September, October, November
        return 'Fall'

# Apply the function to create the 'seasons' column
data['Seasons'] = data['Month'].apply(get_season)

# Drop the 'Month' column if not needed
data.drop('Month', axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
data.head()
```

Now that we have dropped all the categorical data, lets look at the corellation between other variables and Price

```python
data_corr = data.corr()['Price'].map(abs).sort_values(ascending=False)
data_corr
```

```python
sns.set(rc={'figure.figsize':(15, 15)})
#Use the .heatmap method to depict the relationship visually

sns.heatmap(data.corr(), annot=True, annot_kws={"size": 12})
plt.show()
```

From the above analysis, we see that the highest correlation with price is Sqft_living, so lets build a simple linear regression with that

```python
Y = data['Price']
X = data['Sqft_living']
plt.scatter(X,Y)
plt.xlabel("Sqft_living")
plt.ylabel("Price")
plt.show;
```

```python
y = data["Price"]
X_baseline = data[["Sqft_living"]]
```

```python
baseline_model = sm.OLS(y, sm.add_constant(X_baseline))
baseline_results = baseline_model.fit()

print(baseline_results.summary())
```

Our linear regression model is Price = 229.67 * Sqft_living + 4.819e+04.


With a significance level of 0.05, our linear regression model is significant with a p value F statistic of 0.0.


The model explains about 44% of Price as indicated by the R squared.

Both our intercept and our coefficient for Sqft_living are statistically significant.

Our intercept is about 4.819e+04, meaning that a home with 0 square feet living area would cost about $4.819e+4

Our coefficient for Sqft_living is about 229, which means that for each additional square foot living area, we expect the price to increase about $229.

SIMPLE LINEAR REGRESSION VISUALIZATION

```python
#lets plot an actual vs Predicted values.
sm.graphics.plot_fit(baseline_results, "Sqft_living")
plt.show()
```

```python
#lets plot one with a regression line
fig, ax = plt.subplots()
data.plot.scatter(x="Sqft_living", y="Price", label="Data points", ax=ax)
sm.graphics.abline_plot(model_results=baseline_results, label="Regression line", ax=ax, color="black")
ax.legend();
```

```python
# Step 2: Split the data into training and test sets
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Step 3: Train the baseline model and calculate RMSE
# Calculate baseline prediction (e.g., using mean of y_train)
baseline_prediction = np.mean(y_train)# Baseline prediction is the mean
# Create an array of baseline predictions for the test set
Y_baseline_pred = np.full_like(y_test, fill_value=baseline_prediction)
# Calculate RMSE for the baseline model
baseline_rmse = np.sqrt(mean_squared_error(y_test, Y_baseline_pred))

print("Baseline RMSE:", baseline_rmse)
```
